## Listas a usar

```python
import pandas as pd
import numpy as np
import requests # Solicitudes HTTP
import sqlite3 # Manejo de BDD
from bs4 import BeautifulSoup
import re
import os

# Automatización

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time
```

## Automatización 

```python
driver = webdriver.Chrome()
driver.get("https://fbref.com/en/")

try:
    # Hover en "Competitions"
    competiciones = WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.LINK_TEXT, "Competitions"))
    )
    accion = ActionChains(driver)
    accion.move_to_element(competiciones).perform()

    # Click en "La Liga"
    ligas_link = WebDriverWait(driver, 20).until(
        EC.visibility_of_element_located((By.LINK_TEXT, "La Liga"))
    )
    ligas_link.click()
    print("Navegando...")

    # Esperar que aparezcan tablas
    WebDriverWait(driver, 30).until(
        EC.presence_of_all_elements_located((By.TAG_NAME, "table"))
    )
    time.sleep(3)  # extra para que rendericen bien

    # Obtener todas las tablas
    tables = driver.find_elements(By.TAG_NAME, "table")
    print(f"Se encontraron {len(tables)} tablas en la página")

    dataframes = {}
    for i, t in enumerate(tables):
        html = t.get_attribute("outerHTML")
        caption = "tabla_" + str(i)
        if "<caption>" in html:
            caption = html.split("<caption>")[1].split("</caption>")[0]
        df = pd.read_html(html)[0]

        if caption in dataframes:
            caption =caption + f"_{i}"

        dataframes[caption] = df
        print(f"Tabla capturada: {caption}")

finally:
    driver.quit()
```

## Extracción de data

```python
# Filtramos tablas
Data_real = {k: v for k, v in dataframes.items() if not k.startswith("tabla_")}
# Limpiamos titulos de las tablas
Data_limpia = {}
for i, (k, df) in enumerate(Data_real.items(),start=1):
    sin_html = re.sub(r"<.*?>", "", k)
    nombre_limpio = re.sub(r"\s*\d{4}-\d{4}.*Table", "", sin_html).strip()
    nombre_limpio = nombre_limpio.replace("Table", "Overall")
    nombre_limpio = nombre_limpio.replace("Overall_10", "Home/Away")

    if i > 2:
        if re.search(r"_\d+$", k):
            nombre_limpio = re.sub(r"_\d+$", "", nombre_limpio)  # quitar el número
            nombre_limpio = nombre_limpio.replace("Squad", "Opponent")
        else:
            if not nombre_limpio.startswith("Squad"):
                nombre_limpio = "Squad " + nombre_limpio
    
    Data_limpia[nombre_limpio]= df

print("Tablas que contiene la data:\n")
for cont,names in enumerate(Data_limpia.keys(),start=1):
    print(cont,names )
```

```python
# Extracción y depuración
# Información General de la liga en curso
primera_clave = list(Data_limpia.keys())[0]
General = Data_limpia[primera_clave]
General[['Player', 'Goals']] = General['Top Team Scorer'].str.split('-',expand = True)
General['Goals']=pd.to_numeric(General['Goals'])
General = General.drop(columns=['Top Team Scorer'])
# Informacion Home/Away
clave2 = list(Data_limpia.keys())[1]
Home_Away = Data_limpia[clave2]
Home_cols= [col for col in Home_Away.columns if col[0] == 'Home' or col[0] in ['Unnamed: 0_level_0', 'Unnamed: 1_level_0']]
Home = Home_Away[Home_cols].copy()
Home.columns = [
    "Rank" if col[0] == "Unnamed: 0_level_0" 
    else col[1]  # Para las columnas Home
    for col in Home.columns
]
Away_cols= [col for col in Home_Away.columns if col[0]== 'Away' or col[0] in ['Unnamed: 0_level_0', 'Unnamed: 1_level_0']]
Away = Home_Away[Away_cols].copy()
Away.columns=[
    "Rank" if col[0]== "Unnamed: 1_level_0"
    else col[1]
    for col in Away.columns
]
# Estadisticas Estandar Equipo
clave3 = list(Data_limpia.keys())[2]
Equipo_Estadisticas = Data_limpia[clave3].copy()
print(Equipo_Estadisticas.dtypes)
```
